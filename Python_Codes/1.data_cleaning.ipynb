```python
columns_to_drop = [
    "Customer Email",
    "Customer Fname",
    "Customer Lname",
    "Customer Password",
    "Customer Street",
    "Customer Zipcode",
    "Latitude",
    "Longitude",
    "Product Image",
    "Product Description",
    "Order Zipcode"
]
```

```python
# Creating a new customer_table from existing table.
customer_cols = [
    "Customer Id",
    "Customer Fname",
    "Customer Lname",
    "Customer City",
    "Customer State",
    "Customer Country",
    "Customer Segment",
    "Customer Zipcode"
]
```

```python
customer_table = df[customer_cols].drop_duplicates()
print(customer_table.shape)
customer_table.head()
```

```python
# Creating a new product_table from existing table.
product_cols = [
    "Product Card Id",
    "Product Name",
    "Category Id",
    "Category Name",
    "Department Id",
    "Department Name",
    "Product Price"
]
product_table = df[product_cols].drop_duplicates()
print(product_table.shape)
product_table.head()
```

```python
# Creating a new order_table from existing table.
order_cols = [
    "Order Id",
    "order date (DateOrders)",
    "Order Status",
    "Order City",
    "Order State",
    "Order Country",
    "Order Region",
    "Market"
]
order_table = df[order_cols].drop_duplicates()
print(order_table.shape)
order_table.head()

```

```python
# Creating a new shipping_table from existing table.
shipping_cols = [
    "Order Id",
    "Shipping Mode",
    "Delivery Status",
    "Late_delivery_risk",
    "Days for shipping (real)",
    "Days for shipment (scheduled)",
    "shipping date (DateOrders)"
]
shipping_table = df[shipping_cols].drop_duplicates()
print(shipping_table.shape)
shipping_table.head()
```

```python
# Creating a new sales_table from existing table.
sales_cols = [
    "Order Id",
    "Customer Id",
    "Product Card Id",
    "Sales",
    "Order Item Quantity",
    "Benefit per order",
    "Order Profit Per Order"
]
sales_table = df[sales_cols].drop_duplicates()
print(sales_table.shape)
sales_table.head()
```

```python
customer_table.to_csv("customers.csv", index=False)

product_table.to_csv("products.csv", index=False)

order_table.to_csv("orders.csv", index=False)

shipping_table.to_csv("shipping.csv", index=False)

sales_table.to_csv("sales_fact.csv", index=False)
```
